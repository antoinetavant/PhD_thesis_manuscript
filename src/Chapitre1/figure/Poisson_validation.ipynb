```python
# Scientific libraries
import numpy as np
import scipy


# Graphic libraries

import matplotlib.pyplot as plt
%matplotlib notebook

plt.style.use("presentation")

plt.rcParams["figure.figsize"] = (4, 3)
plt.rcParams["font.size"] = 10
```

```python
x_positons = np.linspace(0, 1, 1000)
xmin = 0.475
xmax = 0.525
mask_diel = ( x_positons < xmax) & ( x_positons > xmin)

E = np.zeros_like(x_positons)
E[mask_diel] = 1.15
```

```python
plt.figure(figsize=(4, 3))

plt.plot(x_positons, E-0.115/2, label="Solution")
plt.plot(x_positons, np.ones_like(x_positons)*1.15, "k--", label="$1.15$ V/m")

plt.ylabel("Electric field [V/m]", fontsize=12)
plt.xlabel("Position [m]", fontsize=12)

plt.legend()
plt.savefig("validation_dielectric_field.pdf")
```

```python
phi = - np.cumsum(E-0.115/2)*x_positons[1]

plt.figure()

plt.plot(x_positons, phi, label="Solution")

plt.ylabel("Potential field [V]")
plt.xlabel("Position [m]")
plt.savefig("validation_dielectric_potential.pdf")
```
