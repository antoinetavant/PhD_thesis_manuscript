```python
def get_2Dcut(path):
    F = field(path)
    Er = F.return_cut2d("Ej(2)", "x", 0.5)
    return Er

def get_radial_evol(path, pos=0):
    F = field(path)
    Er = F.return_cut2d("Ej(2)", "y", pos)
    return Er

def print_2Dcut(Er, path, case=1):
    F = field(path)
    
    
    plt.figure(figsize=(4, 3))
    F.definecoords()
    
    y_lim = F._nT*F._dT*F._Na*1e6
    if case == 1:
        x_lim = F._Ly*100
        vlim = np.abs(Er).max()
        vmin = -vlim
        vmax = vlim
        cmap = "seismic"
    elif case==2:
        x_lim = F._Lx*100
        vmin = Er.min()
        vmax = Er.max()
        cmap = "plasma"
        
    plt.imshow(Er*1e-3, cmap=cmap, vmin=vmin*1e-3, vmax=vmax*1e-3, extent=(0, x_lim, 0, y_lim),
              )
    
    if case==1:
        plt.xlabel("Radial position [cm]", fontsize=12)
        
    elif case==2:
        plt.xlabel("Azimuthal position [cm]", fontsize=12)
    
    
    plt.ylabel("Time [$\\mu$s]", fontsize=12)
    cb = plt.colorbar()
    cb.ax.set_ylabel("$E_r$ [kV/m]", fontsize=12)
    
    return plt.gca()
```
